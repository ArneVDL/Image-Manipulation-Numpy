```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
loc_input_img = os.path.join('..', 'data', 'input', 'download.png')
im = np.array(Image.open(loc_input_img))
```

```python
repeated_image = np.tile(im, (3, 8, 1))
```

```python
# Display the original and repeated images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(im)

plt.subplot(1, 2, 2)
plt.title('Repeated Image')
plt.imshow(repeated_image)
plt.show()
```

Creating the mirrored images before tiling

```python
mirrored_image_horizontal = np.flip(im,axis=1)
mirrored_image_vertical = np.flip(im,axis=0)
flipped_both = np.flip(np.flip(im,axis=1),axis=0)
```

```python
plt.imshow(mirrored_image_horizontal)
```

```python
plt.imshow(mirrored_image_vertical)
```

```python
plt.imshow(flipped_both)
```

```python
repeated_image_org = np.tile(im, (1, 6, 1))
repeated_image_hor = np.tile(mirrored_image_horizontal, (1, 6, 1))
repeated_image_vert = np.tile(mirrored_image_vertical, (1, 6, 1))
repeated_image_both = np.tile(flipped_both, (1, 6, 1))
images_array = np.concatenate([repeated_image_org, repeated_image_hor, repeated_image_vert,repeated_image_both], axis=0)
plt.imshow(images_array)

```

```python
red_pixels_only = im.copy()
red_pixels_only[:, :, 1] = 0
red_pixels_only[:, :, 2] = 0
plt.title('Red Pixels Only')
plt.imshow(red_pixels_only)
```

```python
blue_pixels_only = im.copy()
blue_pixels_only[:, :, 1] = 0
blue_pixels_only[:, :, 0] = 0
plt.title('Blue Pixels Only')
plt.imshow(blue_pixels_only)
```

```python
green_pixels_only = im.copy()
green_pixels_only[:, :, 2] = 0
green_pixels_only[:, :, 0] = 0
plt.title('Green Pixels Only')
plt.imshow(green_pixels_only)
```

```python
blue = np.tile(blue_pixels_only, (1, 4, 1))
plt.imshow(blue)
```

```python
red = np.tile(red_pixels_only, (2, 1, 1))
plt.imshow(red)
```

```python
green = np.tile(green_pixels_only, (1, 4, 1))
plt.imshow(green)
```

```python
doubled_size_image = np.repeat(np.repeat(im, 2, axis=0), 2, axis=1)
plt.imshow(doubled_size_image)
```

```python
row2 = np.concatenate([red, doubled_size_image,red], axis=1)
plt.imshow(row2)
```

```python
final = np.concatenate([blue, row2,green], axis=0)
plt.imshow(final)
```

```python
from scipy import ndimage
rotated_image = ndimage.rotate(im, 45, reshape=False)
plt.imshow(rotated_image)
```

```python
black_and_white_image = np.mean(im, axis=-1)
plt.imshow(black_and_white_image, cmap='gray')
```
